```python
import numpy as np
import pandas as pd
import matplotlib as mp 
import seaborn as sns

df = pd.read_csv('house_pricing.csv')
```

```python
df.columns.to_list()
```

```python
filtre = df[(df['LotArea'] < 20000) & (df['SalePrice'] < 500000)]
sns.relplot(x='LotArea', y='SalePrice', data=filtre)
```

```python
filtre2 = df[(df['LotFrontage']< 200) & (df['LotArea']< 100000) ] 
sns.relplot(x='LotFrontage', y='LotArea', data=filtre2)
```
